# Sistema de recomendacion de musica

```python
%matplotlib inline
# importar paquetes necesarios
import pandas
from sklearn.cross_validation import train_test_split
import numpy as np
import time
from sklearn.externals import joblib
import Recommenders as Recommenders
import Evaluation as Evaluation
print("Librerias importadas")
```

# Cargar base de datos de musica

```python
# descargar base de datos de la web
triplets_file = 'https://static.turi.com/datasets/millionsong/10000.txt'
songs_metadata_file = 'https://static.turi.com/datasets/millionsong/song_data.csv'

song_df_1 = pandas.read_table(triplets_file,header=None)
song_df_1.columns = ['user_id', 'song_id', 'listen_count']

# leer metadata de canciones
song_df_2 =  pandas.read_csv(songs_metadata_file)

#Merge the two dataframes above to create input dataframe for recommender systems
song_df = pandas.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on="song_id", how="left") 
```

# Exploracion de datos


```python
# enseñar parte superior de la base de datos
song_df.head(10)
```

```python
# explorar longitud de base de datos
len(song_df)
```

```python
# crear un subconjunto de la base de datos
song_df = song_df.head(10000)
song_df['song'] = song_df['title'].map(str) + " - " + song_df['artist_name']
```

```python
# enseñar las canciones mas populares en la base de datos
song_grouped = song_df.groupby(['song']).agg({'listen_count': 'count'}).reset_index()
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage']  = song_grouped['listen_count'].div(grouped_sum)*100
song_grouped.sort_values(['listen_count', 'song'], ascending = [0,1])
```

```python
# numero de usuarios en la base de datos
users = song_df['user_id'].unique()
len(users)
```

```python
# numero de canciones unicas en la base de datos
songs = song_df['song'].unique()
len(songs)
```

# Crear un sistema de recomendacion de canciones


```python
train_data, test_data = train_test_split(song_df, test_size = 0.20, random_state=0)
print(train_data.head(5))
```

# Crear una instancia de popularidad basdo en clase de recomendacion

```python
pm = Recommenders.popularity_recommender_py()
pm.create(train_data, 'user_id', 'song')
# usar el modelo de popularidad para hacer predicciones
user_id = users[5]
pm.recommend(user_id)
```

# Crear una instancia de recomendacion basado en similaridad

```python
is_model = Recommenders.item_similarity_recommender_py()
is_model.create(train_data, 'user_id', 'song')
```

### Use the personalized model to make some song recommendations

```python
# Usar el modelo personalizado para hacer recomendacion de canciones
# enseñar las canciones para el usuario de la base de datos de entrenamiento
user_id = users[5]
user_items = is_model.get_user_items(user_id)
#
print("------------------------------------------------------------------------------------")
print("Base de datos de entrenamiento para el id del usuario: %s:" % user_id)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_item)

print("----------------------------------------------------------------------")
print("Proceso de recomendacion:")
print("----------------------------------------------------------------------")

# cancion recomendada para el usuario con modelo personalizado
is_model.recommend(user_id)
```

```python
start = time.time()

# definir el porcentaje de usuarios para usar precision y recall
user_sample = 0.05

# instanciar la clase precision_recall_calculator
pr = Evaluation.precision_recall_calculator(test_data, train_data, pm, is_model)

# llamar el metodo para calcular precision y recall
(pm_avg_precision_list, pm_avg_recall_list, ism_avg_precision_list, ism_avg_recall_list) = pr.calculate_measures(user_sample)

end = time.time()
print(end - start)
```
